```python
# Create a list called things with three string elements

things = ["mozzarella", "cinderella", "salmonella"]
print(things)
```

```python
# Capitalize the element that refers to a person (Cinderella)

things[1] = things[1].capitalize()   # Cinderella is a person
print(things)
```

```python
# Make the cheesy element (mozzarella) all uppercase

things[0] = things[0].upper()        # mozzarella is the cheese
print(things)
```

```python
# Delete the disease element (salmonella) from the list

things.remove("salmonella")          # salmonella is the disease
print(things)
print("    🏆 Nobel Prize collected!")
```

```python
#Define a generator function that yields odd numbers from range(10)
# Use a for loop to find and print the third value returned

def get_odds():
    """Generator that yields odd numbers from range(10)."""
    for number in range(10):
        if number % 2 != 0:   # Check if the number is odd
            yield number       # yield pauses and returns the value

# Use a for loop to reach the third odd number
count = 0
for value in get_odds():
    count += 1
    if count == 3:             # Stop at the third odd number
        print (value)
        break
```
